```python
import mechanicalsoup as ms
import bs4 as bs
import re
from win10toast import ToastNotifier
import pandas as pd
```

```python
#
browser = ms.StatefulBrowser()
df = pd.DataFrame()
df = pd.read_csv(r'C:\Users\hm007677\OneDrive\Odilon Behrens\Banco de dados\OSPython.csv')
toaster = ToastNotifier()
url = "http://crm.odilon.pbh/users/sign_in"
```

```python
login_page = browser.get(url)
login_html = login_page.soup
form = login_html.select("form")[0]
form.select("input")[2]["value"]= "lucas@hob.pbh"
form.select("input")[3]["value"] = "95123698741"
profiles_page = browser.submit(form,login_page.url)
```

```python
profiles_page.url
```

```python
OS = profiles_page.soup.get_text().replace("\n","").split("#")
```

```python
listaOS = []
item=0
while (OS[item+1].split("-")[0] > df['OS'][0].split('-')[0] and item<25):
#for item in (range(len(OS)-2)):#O último item do array OS corresponde à página seguinte e não possui dados, logo é desconsiderado
    itemOS = []  #Variável para armazenar a linha de cada OS. Reiniciada a cada ciclo do for
    ticket = browser.get("http://crm.odilon.pbh/engenharia-clinica---hob-12/tickets/"+OS[item+1].split("-")[0])
    if ticket.soup.find_all(attrs={'selected':'selected'}): #Confere se a OS possui setor selecionado
        setor = ticket.soup.find_all(attrs={'selected':'selected'})[0].get_text().split('') #Lê nome do setor e guarda na variável
    else:
        setor = ''
    comentario = ticket.soup.find('textarea').get_text() #Lê o comentário adicionado pelo usuário
    #detalhes = OS[item+1].split("-")
    itemOS = [OS[item+1],setor,comentario] #Cria a linha da OS com todos os dados 
    listaOS.append(itemOS) 
    df = df.append(pd.Series(itemOS,index=['OS','Setor','Comentário']),ignore_index=True) #Adiciona a linha ao DataFrame
    print(item)
    item+=1
df = df.sort_values(by='OS',ascending=False).reset_index(drop=True)
df
df.to_csv(r'C:\Users\hm007677\OneDrive\Odilon Behrens\Banco de dados\OSPython.csv',index = False)
```
